```python
import cv2 as cv
img01 = cv.imread('LogoUNIR.png')
img02 = cv.imread('LogoUNIR.png',cv.IMREAD_GRAYSCALE)
alto1, ancho1, canales1 = img01.shape
alto2, ancho2 = img02.shape

print(alto1, ancho1, canales1)
print(alto2, ancho2)

cv.imshow("img01 - Colores",img01)
cv.imshow("img02 - BN",img02)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
from matplotlib import pyplot as plt 

def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

plt.imshow(cv.cvtColor(img01, cv.COLOR_BGR2RGB))
plt.title('imagen')
plt.show()
```
