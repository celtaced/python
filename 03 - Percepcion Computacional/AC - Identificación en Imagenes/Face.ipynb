# Detección de Rostros
Algoritmo para detección de rostros en una imágen haciendo uso de clasificadores Haar

(Aun pendiente de ajustar bien los parametros de detección)

```python
# Carga de Librerías
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Carga de imagen
imagePath = 'Test02.png'
img = cv.imread(imagePath)
img.shape

plt.figure(figsize=(10,5))
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Imagen Original")
plt.axis('off')
```

```python
gray_image = cv.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.shape

plt.figure(figsize=(10,5))
plt.imshow(cv.cvtColor(gray_image, cv.COLOR_BGR2RGB))
plt.title("Imagen Original")
plt.axis('off')
```

## Preprocesamiento de la Imagen

```python
# Aumento de contraste
equalized_image = cv.equalizeHist(gray_image)

# Reducción de ruido
blurred_image = cv.GaussianBlur(equalized_image, (5, 5), 0)
```

## Clasificadores Haar:
- Son clasificadores tradicionales de visión por computadora que se basan en características extraídas manualmente de las imágenes (bordes, formas, etc.).
- Son rápidos y eficientes, pero pueden tener limitaciones en cuanto a precisión y capacidad para detectar rostros en diferentes poses o condiciones de iluminación.

```python
# Carga el clasificador preentrenado de Haar para la detección de rostros
face_classifier = cv.CascadeClassifier(
    cv.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
# Detecta los rostros en la imagen
faces = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.008, minNeighbors=3, minSize=(100, 100)
)
```

```python
# Dibuja rectángulos alrededor de los rostros detectados
for (x, y, w, h) in faces:
    cv.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img_rgb)
plt.axis('off')
```
