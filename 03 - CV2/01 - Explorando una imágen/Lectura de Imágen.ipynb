```python
# Bibliotecas a utilizar.
from matplotlib import pyplot as plt
import cv2 as cv

# Imagen a utilizar
img = 'test.jpg'
```

```python
# Obtener imagen
img_color =  cv.imread(img) # Por defecto OpenCV leerá la imágen en formato BGR
img_gray = cv.imread(img, cv.IMREAD_GRAYSCALE)
Al1, An1, Can1 = img_color.shape
print(
        'Color:\nimg_color_alto:',Al1, 
        '\nimg_color_ancho:',An1, 
        '\nimg_color_canales:', Can1
      )

Al2, An2 = img_gray.shape
print(
        '\nGrises:\nimg_gray_alto:',Al2, 
        '\nimg_gray_ancho:',An2
      )
```

```python
# Preview de imagenes cargadas
cv.imshow('Imagen a Color', img_color)
cv.imshow('Imagen en Grises', img_gray)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Mostrar las imagenes con Matplotlib
plt.imshow(cv.cvtColor(img_color, cv.COLOR_BGR2RGB)) # Convertimos BGR a RGB
plt.title("Imagen")
plt.show()
```

# Como se lee una imágen con OpenCV


```python
# Seleccionar un sector de la imagen
def showImg(ImagenOriginal, a=25, b=225, c=25, d=225):
    img_test = ImagenOriginal
    SmallImg =  img_test[a:b,c:d]
    try:
        cv.imshow('Imagen Original', img_test)
        cv.imshow('Seccion de imagen', SmallImg)
        cv.waitKey(0)
        cv.destroyAllWindows()
    except:
        return 'Error'
    

showImg(img_gray, 25, 255, 200, 4000)
```

# Leer un pixel

```python
# Imagen a colores
pixel_xycolor = img_color[18,160]
print("Pixel color [BGR]",pixel_xycolor)

# Imagen en escala de grises
pixel_xygray=img_gray[18,160]
print("Pixel grises",pixel_xygray)
```
